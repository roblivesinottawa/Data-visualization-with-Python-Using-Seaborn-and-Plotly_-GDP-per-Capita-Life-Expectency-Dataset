```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('2019.csv')
```

```python
data.head()
```

Ditribution plot

```python
sns.distplot(data['GDP per capita']) # kde=False if you want not show it
```

```python
sns.distplot(data['GDP per capita'], kde = False , bins = 30)
```

```python
sns.jointplot(x=data['GDP per capita'], y= data['Healthy life expectancy'],data=data)# two ditribution
```

```python
sns.jointplot(x=data['GDP per capita'], y= data['Healthy life expectancy'],data=data,kind='reg')# linear regression
```

```python
sns.jointplot(x=data['GDP per capita'], y= data['Healthy life expectancy'],data=data,kind='hex')
```

```python
sns.pairplot(data)#relationship entire datasets
```

categorical plot

```python
sns.barplot(x=data['Country or region'].head(3),y=data['GDP per capita'],data=data)
```

```python
sns.boxplot(x=data['Country or region'].head(2),y=data['GDP per capita'].head(2) ,data=data)
```

```python
sns.violinplot(x=data['Country or region'].head(2),y=data['GDP per capita'] ,data=data,palette='rainbow')
```

matrix plot

```python
 # Matrix form for correlation data
data_select = data[['GDP per capita','Social support','Healthy life expectancy','Perceptions of corruption']]
data_select.corr()
```

```python
sns.heatmap(data_select.corr(), cmap='coolwarm')
```

```python
sns.heatmap(data_select.corr(), cmap='coolwarm', annot=True)
```

pairgrid

```python
# Just the Grid
sns.PairGrid(data_select)
```

```python
# Then you map to the grid
g = sns.PairGrid(data_select)
g.map(plt.scatter)
```
