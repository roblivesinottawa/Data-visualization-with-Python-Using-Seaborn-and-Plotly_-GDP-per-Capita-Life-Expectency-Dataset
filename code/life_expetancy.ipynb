```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("darkgrid")
```

# PLOTTING GRAPHS USING SEABORN 

```python
df = pd.read_csv('../data/2019.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['Country or region'][31]
```

```python
# exploratory data analysis
sns.displot(df['GDP per capita'], kde=False)
```

```python
sns.displot(df['GDP per capita'], kde=False, color='r', bins=30)
```

```python
# two distribution
sns.jointplot(x=df['GDP per capita'], y=df['Healthy life expectancy'], data=df, color='r')
```

```python
# linear regression
sns.jointplot(x=df['GDP per capita'], y=df['Healthy life expectancy'], data=df, kind='reg', color='b')
```

```python
sns.jointplot(x=df['GDP per capita'], y=df['Healthy life expectancy'], data=df, kind='hex')
```

```python
# get the relationships in the dataset
sns.pairplot(df)
```

categorical plot

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=df['Country or region'].head(7), y=df['GDP per capita'], data=df, color=sns.color_palette()[0])
```

```python
sns.boxplot(x=df['Country or region'].head(5), y=df['GDP per capita'], data=df)
```

```python
sns.violinplot(x=df['Country or region'].head(5), y=df['GDP per capita'], data=df, palette=sns.color_palette())
```

matrix plot

```python
data_select = df[['GDP per capita', 'Social support', 'Healthy life expectancy', 'Perceptions of corruption']]
data_select.corr()
```

```python
sns.heatmap(data_select.corr(), cmap='viridis', annot=True)
```

pairgrid

```python
sns.PairGrid(data_select)
```

```python
# map it to the grid
g = sns.PairGrid(data_select)
g.map(plt.scatter)
```
